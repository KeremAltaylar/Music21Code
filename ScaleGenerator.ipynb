```python
def ScaleMakerWithKey(root, key_signature,families, time_signature, tempo_mark,names):
    Scales = stream.Score()
    Scales.insert(0, metadata.Metadata())
    Scales.metadata.title = 'SCALE FAMILIES'
    for number in range(0,len(families)):
        sC = stream.Part(id=str(number))
        measure = stream.Measure(id=str(number))
        # measure.append(metadata.Metadata()) # insert(0,metadata)
        # tb = text.TextBox(str(names[number]), 250, 100)
        # tb.style.fontSize = 40
        # tb.style.alignVertical = 'bottom'   
        measure.append(tb)
        measure.append(key.Key(key_signature))
        measure.append(meter.TimeSignature(time_signature))
        #measure.append(tempo.MetronomeMark(number=tempo_mark))
        sC.append(measure)
        for semitones in families[number]:
            tempInt = interval.Interval(semitones)
            tempInt.noteStart = note.Note(root)
            name = note.Lyric(tempInt.name)
            name.style.color = 'blue'
            name.style.size = 3
            Note = tempInt.noteEnd
            Note.lyrics.append(name)
            Note.addLyric(Note.pitch.unicodeNameWithOctave)
            Note.addLyric(Note.pitch.midi)
            Note.addLyric(round(Note.pitch.frequency,2))
            measure.append(Note)
            #print(semitones, tempInt.niceName)
        Scales.append(sC)
    Scales.show()
    Scales.show("text")
    #sC.show()
    #sC.show('text')
    return Scales

#ScaleMakerWithKey('C4',48,'C', 4/4, 120)
```
