```python
import random
import music21
from music21 import stream, note, meter, key, tempo, roman, interval, chord,scale, clef,layout,duration,instrument,metadata
from itertools import combinations
from music21.figuredBass import possibility
import pathlib
from icecream import ic
music21.environment.set('musescoreDirectPNGPath', 
                        pathlib.PosixPath('/Applications/MuseScore 4.app/Contents/MacOS/mscore'))

## Self notes:
# * There should be a determinant of how many voices will be at the moment for example 3 then 2 then 4 for howw many chords
# * In the selected key determine is it major or minor
# * Upon the determination, select useful scales not only scales but modes

#current_chord = roman.RomanNumeral('ii7', 'c')


def create_score(part_number,scoreKey,name,tempo):
  score = stream.Score(id=name)
  score.append(music21.tempo.MetronomeMark(number=tempo))
  score.insert(0, metadata.Metadata())
  score.metadata.title = name
  score.metadata.composer = 'Kerem Altaylar'


  for part_index in range(part_number):
    
    if part_index == 2:
      part3 = stream.Part(id=f"Part{part_index + 1}")  # Use f-string for dynamic IDs
      part3.append(music21.tempo.MetronomeMark(number=tempo))

      part3.append(key.Key(scoreKey))
      part3.append(meter.TimeSignature('4/4'))
      score.insert(part_index,part3)
      flute = instrument.Flute()
      part3.insert(0, flute)

    elif part_index == 3:
      part4 = stream.Part(id=f"Part{part_index + 1}")  # Use f-string for dynamic IDs
      part4.append(key.Key(scoreKey))
      part4.append(meter.TimeSignature('4/4'))
      score.insert(part_index,part4)  
      trombone = instrument.Viola()
      part4.insert(0, trombone) 

    elif part_index == 0:
      part1 = stream.Part(id=f"Part{part_index + 1}")  # Use f-string for dynamic IDs
      part1.append(key.Key(scoreKey))
      part1.append(meter.TimeSignature('4/4'))
      score.insert(part_index,part1) 
      violin = instrument.Violin()
      part1.insert(0, violin)

    elif part_index == 1:
      part2 = stream.Part(id=f"Part{part_index + 1}")  # Use f-string for dynamic IDs
      part2.append(key.Key(scoreKey))
      part2.append(meter.TimeSignature('4/4'))
      score.insert(part_index,part2)
      harp = instrument.Harp()
      part2.insert(0, harp)
    elif part_index == 4:
      part5 = stream.Part(id=f"Part{part_index + 1}")  # Use f-string for dynamic IDs
      part5.append(key.Key(scoreKey))
      part5.append(meter.TimeSignature('4/4'))
      score.insert(part_index,part5)
      organ = instrument.Violoncello()
      part5.insert(0, organ)
    elif part_index == 5:
      part6 = stream.Part(id=f"Part{part_index + 1}")  # Use f-string for dynamic IDs
      part6.append(clef.BassClef())
      part6.append(key.Key(scoreKey))
      part6.append(meter.TimeSignature('4/4'))
      score.insert(part_index,part6)
      cello = instrument.Violoncello()
      part6.insert(0,cello)
    elif part_index == 6:
      part8 = stream.Part(id=f"Part{part_index + 1}")  # Use f-string for dynamic IDs
      part8.append(key.Key(scoreKey))
      part8.append(meter.TimeSignature('4/4'))
      score.insert(part_index,part8)
      piano = instrument.Piano()
      part8.insert(0,piano)
  return score

def generate_modes_with_notes(base_key):
    if base_key.mode == 'major':
        scale_obj = scale.MajorScale(base_key.tonic)
    elif base_key.mode == 'minor':
        scale_obj = scale.MinorScale(base_key.tonic)
    else:
        raise ValueError("Unsupported key mode. Use 'major' or 'minor'.")
    
    pitches = scale_obj.getPitches()
    
    # Mode names and their sequence
    mode_names = ['Ionian', 'Dorian', 'Phrygian', 'Lydian', 'Mixolydian', 'Aeolian', 'Locrian']
    modes = []
    
    for i, mode_name in enumerate(mode_names):
        mode_pitches = pitches[i:] + pitches[:i]  # Rotate the scale
        mode_notes = [p.nameWithOctave for p in mode_pitches]
        mode_tonic = mode_pitches[0].name  # Mode's tonic note
        modes.append((mode_tonic, mode_name, mode_notes))
    
    return modes

def display_modes_for_key(input_key):
    base_key = key.Key(input_key)
    modes = generate_modes_with_notes(base_key)
    
    output = []
    output.append(f"Key: {base_key.tonic} {base_key.mode}")
    output.append("Usable:")
    
    for tonic, mode_name, notes in modes:
        output.append(f"- {tonic} {mode_name}: {', '.join(notes)}")
    
    return "\n".join(output)

def generate_chord_progression(length=32, key_signature= "Eb", time_signature='4/4', tempo_mark=360,nameS='',partNumber = 6):
    # Create a stream for the melody and chords
    score = create_score(partNumber,key_signature,nameS,tempo_mark)
    key_obj = key.Key(key_signature)
    scales= generate_modes_with_notes(key_obj)
    print(scales)

    #sc1 = scale.MinorScale(key_signature)
    sc1 = scale.MajorScale(key_signature)
    sc1Ext = scale.ConcreteScale(
    pitches=[str(p) for p in sc1.getPitches('c1', 'c8', direction=scale.Direction.ASCENDING)])
    sc1High = scale.ConcreteScale(
    pitches=[str(p) for p in sc1.getPitches('c4', 'c7', direction=scale.Direction.ASCENDING)])
    sc1Bass = scale.ConcreteScale(
    pitches=[str(p) for p in sc1.getPitches('c1', 'c4', direction=scale.Direction.ASCENDING)])


    # Melody part
 
    # Chords part
    chords_part = stream.Part()
    score.append(chords_part)
    bassClef = clef.BassClef()
    chords_part.append(bassClef)
    # Set the key and time signature
    chords_part.append(key_obj)
    chords_part.append(meter.TimeSignature(time_signature))

    # Voices Part

    mChords = [
    'i', 'i6', 'i64',
    'ii', 'ii43', 'ii65',
    'III', 'III64',
    'iv', 'iv6', 'iv64',
    'v', 'v6', 'v64',
    'VI', 'VI6', 'VI64',
    'VII', 'VII6', 'VII64'
]
    mChordsExt = [
    'i', 'i6', 'i64', 'i7', 'i65', 'i43', 'i9', 'i11',
    'ii', 'ii43', 'ii65', 'ii7', 'ii753', 'ii975', 'ii9', 'ii11',
    'III', 'III64', 'III7', 'III65', 'III43', 'III9', 'III11',
    'iv', 'iv6', 'iv64', 'iv7', 'iv65', 'iv43', 'iv9', 'iv11',
    'v', 'v6', 'v64', 'v7', 'v65', 'v43', 'v9', 'v11',
    'VI', 'VI6', 'VI64', 'VI7', 'VI65', 'VI43', 'VI9', 'VI11',
    'VII', 'VII6', 'VII64', 'VII7', 'VII65', 'VII43', 'VII9', 'VII11'
]

    # Choose the initial chord (I or IV)
    initial_chord_roman = random.choice(mChords)
    cLength = random.choice([4,3])
    print(initial_chord_roman)
    # initial_chord_roman = 'VI7'
    current_chord = roman.RomanNumeral(initial_chord_roman, key_obj,quarterLength=cLength)
    current_chord_pitches = current_chord.pitches

    # Spreead Octaves of first chord
    for pitch in range(0,len(current_chord)-1):
        current_chord[pitch].transpose(12 * (pitch-2), inPlace=True)
    
    chord_member_indexes = list(range(0,len(current_chord_pitches)))
    
    lyric = note.Lyric(roman.romanNumeralFromChord(current_chord, key_obj).figure)
    lyric.style.color = 'red'
    current_chord.lyrics.append(lyric)

    if len(chord_member_indexes) < 4:
        duplicate = current_chord.bass()
        duplicate = duplicate.transpose(12)
        current_chord.add(duplicate)
    
    current_chord_pitches = current_chord.pitches
    chord_member_indexes = list(range(0,len(current_chord_pitches)))

    current_chord_pitches = current_chord.pitches

    
    changeChordsNumbers = list()
    changeChordsIndexes = list()
    
    for cNumber in range(0,length ):
        # Each Chord in the part is iterated
        nextChord = chord.Chord()
        posFifth = False
        # Number of Chord members for each chord that will change for next chord
        number_change_chord = random.randint(1,len(current_chord_pitches)-1)
        changeChordsNumbers.append(number_change_chord)
        chord_member_indexes = list(range(0,len(current_chord_pitches)))
        print("{} {}".format('Chord: ', cNumber+1))
        print("{} {}".format('Current Length of the Chord: ', len(current_chord_pitches)))
        print("{} {}".format('How many chord members will change: ', changeChordsNumbers[cNumber]))
        currentChangeChordIndex = random.choice(list(combinations(chord_member_indexes,changeChordsNumbers[cNumber])))
        changeChordsIndexes.append(currentChangeChordIndex) # deiecek chord yeleri
        print("{} {}".format('Which members of the Chord will change: ',changeChordsIndexes[cNumber])   )
        #Determine new chords members to previous or next note in the key
        chordLength = random.choice([4,2,8,1])
        
        for member in range(len(current_chord_pitches)):
        # Each member of chord is iterated
            if member in changeChordsIndexes[cNumber]:
                write = 'will change'
                # iChange = interval.Interval(random.choice(['m2','m-2']))
                # bPitch = interval.transposePitch(current_chord_pitches[member], iChange)
                dirType = random.choice([scale.Direction.DESCENDING,scale.Direction.ASCENDING])
                if dirType == scale.Direction.DESCENDING:
                    directionMember = 'down'
                if dirType == scale.Direction.ASCENDING:
                    directionMember = 'up'
                changedPitch = sc1Ext.nextPitch(current_chord_pitches[member],dirType)
                nextChord.quarterLength = chordLength
                nextChord.add(changedPitch)
            else: 
                write = 'will not change'
                remainingPitch = current_chord_pitches[member]
                nextChord.quarterLength = chordLength
                nextChord.add(remainingPitch)
                directionMember = ''
            print("{} {} {} {}".format('Member number ',member, write,directionMember))
        # VOICE CONTROL MECHANISM WILL BE HERE
        # !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

 
        for m in range(0,partNumber):
          motiveLength = random.randint(1,4)
          if (m ==5):
            motiveMaker(motiveLength,1,score.parts[m],nextChord,chordLength)
          if (m ==3):
            motiveMaker(motiveLength,1,score.parts[m],nextChord,chordLength)
          motiveMaker(motiveLength,1,score.parts[m],nextChord,chordLength)

        #Look for paralel fifths and octaves
        posFifth = possibility.parallelFifths(current_chord.pitches, nextChord.pitches)
        print("{} {}".format('Paralel Fifths: ',posFifth)   )
        posOctave = possibility.parallelOctaves(current_chord.pitches, nextChord.pitches)
        print("{} {}".format('Paralel Octaves: ',posOctave)   )
        hosFifth = possibility.hiddenFifth(current_chord.pitches, nextChord.pitches)
        print("{} {}".format('Hidden Fifths: ',hosFifth))
        hosOctave = possibility.hiddenOctave(current_chord.pitches, nextChord.pitches)
        print("{} {}".format('Hidden Octaves: ',hosOctave))
        vosOverlap = possibility.voiceOverlap(current_chord.pitches, nextChord.pitches)
        print("{} {}".format('Voice Overlap: ',vosOverlap))
            
        # Duplicate bass an octave higher if the chord is 3 Chord
        if len(chord_member_indexes) == 3:
            duplicate = nextChord[0]
            duplicate = duplicate.transpose(-12)
            nextChord.add(duplicate)
            duplicate1 = nextChord[1]
            duplicate1 = duplicate1.transpose(12)
            nextChord.add(duplicate1)
            duplicate2 = nextChord[2]
            duplicate2 = duplicate2.transpose(12)
            nextChord.add(duplicate2)
        elif len(chord_member_indexes) == 4:
            duplicate = nextChord[0]
            duplicate = duplicate.transpose(12)
            nextChord.add(duplicate)
            duplicate1 = nextChord[1]
            duplicate1 = duplicate1.transpose(-12)
            nextChord.add(duplicate1)
        elif len(chord_member_indexes) == 5:
            duplicate = nextChord.bass()
            duplicate = duplicate.transpose(12)
            nextChord.add(duplicate)

        # Next chord is added to the score!     
        chords_part.append(nextChord)
        
        # Distribution of chor members to parts
  
  #! This is the part adds chords to parts 
        # for part_index in range(part_number):
        #     partNote = note.Note(nextChord[part_index].pitch)
        #     partNote.duration.quarterLength = chordLength
        #     score.parts[part_index].append(partNote)
        # for part in score.parts:
        #     for noteV in part.getElementsByClass(note.Note):
        #         if noteV.duration is None:
        #             noteV.setV(64)

        
        
        #change the colors of roman numerals for paralel Fifths and Octaves
        if posOctave:
            lyric = note.Lyric(roman.romanNumeralFromChord(nextChord, key_obj).figure)
            lyric.style.color = 'blue'
            nextChord.lyrics.append(lyric)
        elif posFifth:
            lyric = note.Lyric(roman.romanNumeralFromChord(nextChord, key_obj).figure)
            lyric.style.color = 'green'
            nextChord.lyrics.append(lyric)
        # Add Roman Numeral
        elif hosFifth:
            lyric = note.Lyric(roman.romanNumeralFromChord(nextChord, key_obj).figure)
            lyric.style.color = 'yellow'
            nextChord.lyrics.append(lyric)
        # Add Roman Numeral
        elif hosOctave:
            lyric = note.Lyric(roman.romanNumeralFromChord(nextChord, key_obj).figure)
            lyric.style.color = 'purple'
            nextChord.lyrics.append(lyric)
        elif vosOverlap:
            lyric = note.Lyric(roman.romanNumeralFromChord(nextChord, key_obj).figure)
            lyric.style.color = 'red'
            nextChord.lyrics.append(lyric)
        # Add Roman Numeral
        else:
            lyric = note.Lyric(roman.romanNumeralFromChord(nextChord, key_obj).figure)
            lyric.style.color = 'black'
            nextChord.lyrics.append(lyric)

        # Replace Current chord pitch with next
        current_chord = nextChord
        current_chord_pitches = nextChord.pitches
    print(sc1Bass.pitches)
    print(sc1High.pitches)
    return score

def tupleBlockMaker(numTuple, notes,qdurations,quant,part,chordL):

  for ind in range(numTuple):
    t = duration.Tuplet(numTuple, quant)
    #t.setDurationType(qdurations)
    d = duration.Duration(qdurations)
    d.appendTuplet(t)

    blockLength = qdurations * quant
    numNote = int(chordL/blockLength)
    #n = note.Note(notes[ind].name)
    for numbN in range(numNote):
      n = note.Note(random.choice(notes).name)
      #n = n.transpose(12)
      n.duration = d
      part.append(n)    #Append many times

  # n = note.Note(notes)
  # n.duration = d
  # part3.repeatAppend(n, numNote)   

def normalBlockMaker(num, notes,qdurations,part):
  for ind in range(num):
    #note1 = note.Note(notes[ind].name)
    indeks = random.randint(1,20)

    if (indeks < 12):
        note1 = note.Note(notes[ind % 3].name)
    elif (indeks > 1):
        note1 = note.Note(random.choice(notes).name)

 

    note1.duration.quarterLength = (qdurations)
    part.append(note1)

  # note1 = note.Note(notes)
  # note1.duration.quarterLength = (qdurations)
  # part3.repeatAppend(note1, num)

def restBlockMaker(num, notes,qdurations,part):
  for ind in range(num):
    #note1 = note.Note(notes[ind].name)
    indeks = random.randint(1,20)

    if (indeks < 12):
        note1 = note.Rest()
    elif (indeks > 1):
        note1 = note.Rest()

    #note1.transpose("P8", inPlace=True)
    note1.duration.quarterLength = (qdurations)
    part.append(note1)

  # note1 = note.Note(notes)
  # note1.duration.quarterLength = (qdurations)
  # part3.repeatAppend(note1, num)

def motiveMaker(bNum,blockNumb,part, pitches,chordL):
  #tuples
  for p in range(1,bNum+1):
    for b in range(1,blockNumb+1):
      indeks = random.randint(1,20)

      if (indeks < 4):
        tupleNumb = random.choice([3])
        normalNumb = random.choice([0.5])
        if (normalNumb == 0.5):
          noteNumb = 2
        elif (normalNumb == 1):
          noteNumb = 4
        elif (normalNumb == 2):
          noteNumb = 8
        #tupleBlockMaker(tupleNumb,pitches,normalNumb,noteNumb,part,chordL)
        normalBlockMaker(noteNumb,pitches,normalNumb,part)
      elif(indeks > 10):
        normalNumb = random.choice([0.5])
        if (normalNumb == 0.5):
          noteNumb = 2
        elif (normalNumb == 1):
          noteNumb = 4
        elif (normalNumb == 2):
          noteNumb = 8
        restBlockMaker(noteNumb,pitches,normalNumb,part)
      else:
        normalNumb = random.choice([0.25,0.5,1,2])
        noteNumb = int(chordL / normalNumb)
        normalBlockMaker(noteNumb,pitches,normalNumb,part)
    
    # Show the random melody with chords
#random_melody_with_chords.show('text')
#random_melody_with_chords.show()



major_keys = [
    'C', 'G', 'D', 'A', 'E', 'B', 'F#', 'C#',  # Sharp keys
    'F', 'Bb', 'Eb', 'Ab', 'Db', 'Gb', 'Cb'    # Flat keys
]

# Minor keys
minor_keys = [
    'a', 'e', 'b', 'f#', 'c#', 'g#', 'd#', 'a#',  # Sharp keys
    'd', 'g', 'c', 'f', 'bb', 'eb', 'ab'          # Flat keys
]

# Combined list of all keys
all_keys = major_keys + minor_keys

for i in range(1,2):
  title = 'Generations no:'
  keyS = random.choice(all_keys)
  name = "{}{}{}{}".format(title,str(i),' in ', str(keyS))
  random_melody_with_chords = generate_chord_progression(20,keyS, '4/4', 60, name,6)
  path = "/Users/metalblade/Desktop/Archieve/CODE/Python/Music21/Output/Generations"
  endpath = ("{}{}{}{}{}{}".format(path,' no: ',str(i),' in ',str(keyS),'.xml'))
  print(endpath)
  random_melody_with_chords.write("musicxml", endpath)
  random_melody_with_chords.show()
  random_melody_with_chords.show('text')
  

  


#random_melody_with_chords.write("musicxml", "/Users/metalblade/Desktop/Archieve/CODE/Python/Music21/Output/Score1.xml")

# Partlarn iine susacaklar ekilde restler eklenicek
# blocklarn iine susacaklar ekilde restler eklenecek
# Motive generator iin Chord tone, non-chord tone, chromatic seenekleri eklenecek
```
