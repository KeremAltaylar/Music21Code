```python
from music21 import stream, instrument, clef,note, meter,bar,metadata, key, tempo,scale,pitch
```

```python
s = stream.Measure()
ts1 = meter.TimeSignature('3/4')
s.insert(0, ts1)
s.insert(0, key.KeySignature(2))
s.insert(0, clef.TrebleClef())
s.insert(0, note.Note('C#4'))
s.insert(1, note.Note('D#4'))
e = note.Note('E4')
s.append(e)
s.show()
print(e.offset)
print(s[4])
s.getElementsByOffset(2.0)[0]
```

```python
s.show('text')
print((s[0], s[1], s[2]))
print((s[0].classSortOrder, s[1].classSortOrder, s[2].classSortOrder))
print((bar.Barline.classSortOrder, instrument.Instrument.classSortOrder, metadata.Metadata.classSortOrder))
```

```python
ts2 = meter.TimeSignature('6/8')
s.insert(0, ts2)
s.show('text')
```

```python
d = note.Note('D')
d.priority = -10
eis = note.Note('E#')
eis.priority = 10
s.insert(1.0, d)
s.insert(2.0, eis)
s.show('text')
```

```python
d.priority = 20
s.remove(d)
s.insert(1.0, d)
s.show('text')
```

```python
ts2.priority = -5
s.remove(ts2)
s.insert(0.0, ts2)
s.show('text')
s.show()
```

```python
for el in s.getElementsByOffset(0.0):
    el.priority = el.classSortOrder

ts2.priority = 3  # between KeySignature (priority = 2) and TimeSignature (priority = 4)
s.show('text')
s.show()
```
