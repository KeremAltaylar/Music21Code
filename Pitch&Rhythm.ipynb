```python
# Duration Object
#  “whole”, “half”, “quarter”, “eighth”, “16th”, “32nd”, “64th”
#  "breve" , "longa" , "maxima" , "128th" , "265th" , "2048th"
halfDuration = duration.Duration('half')
dottedQuarter = duration.Duration(1.5)
dottedQuarter.quarterLength
halfDuration.quarterLength
halfDuration.type
dottedQuarter.type
halfDuration.dots
dottedQuarter.dots
# Properties of Duration (A special attribute)
# Quarter Note Length is very important, 1/3 is eight note triplet
dottedQuarter.dots = 2
print(dottedQuarter.quarterLength)
dottedQuarter.dots = 3
print(dottedQuarter.quarterLength)
dottedQuarter.dots = 4
print(dottedQuarter.quarterLength)
dottedQuarter.quarterLength = 0.25
print(dottedQuarter.type)

```
