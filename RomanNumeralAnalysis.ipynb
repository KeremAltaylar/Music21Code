```python
from music21 import *
f = chord.Chord('F4 C5 A5')
kf = key.Key('F')
sf = stream.Measure([kf, f])
sf.show()
kc = key.Key('C')
sc = stream.Part([kc, f])
sc.show()
kb = key.Key('B')
sb = stream.Part([kb, f])
sb.show()
```

```python
rf = roman.romanNumeralFromChord(f, kf)
rc = roman.romanNumeralFromChord(f, kc)
rb = roman.romanNumeralFromChord(f, kb)
print(rb,rc,rf)
print((rf.figure, rc.figure, rb.figure))
print(rf.figureAndKey)
print((rf.scaleDegree, rc.scaleDegree, rb.scaleDegree))
print(rf.scaleDegreeWithAlteration)
print(rb.scaleDegreeWithAlteration)
print((rf.functionalityScore, rc.functionalityScore, rb.functionalityScore))
s = stream.Measure()
s.insert(0, rf)
s.show()
```
