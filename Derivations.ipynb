```python
from music21 import *
s = corpus.parse('mozart/k80', 1)
s.id = 'mozartK80'
s.derivation
```

```python
sExcerpt = s.measures(1, 4)
sExcerpt.id = 'excerpt'
sExcerpt.show()
```

```python
print(sExcerpt.derivation)
print(sExcerpt.derivation.client)
print(sExcerpt.derivation.origin)
print(sExcerpt.derivation.method)
sTransposed = sExcerpt.transpose('P4')
sTransposed.show()


```

```python
sTransposed.id = 'transposed'
print(sTransposed.derivation)
print(sTransposed.derivation.client)
print(sTransposed.derivation.origin)
print(sTransposed.derivation.method)
print(sTransposed.derivation.rootDerivation)
print(sTransposed.derivation.origin.derivation.origin)

```

```python
def makeLouder(s):
    for dyn in s.recurse().getElementsByClass('Dynamic'):
        dyn.value = 'f'

makeLouder(sTransposed)
sTransposed.show()
sExcerpt.show()
```

```python
for s in sTransposed.derivation.chain():
    makeLouder(s)

sExcerpt.show()
```
